## 4.1 üöª Propor√ß√£o de estudantes por g√™nero

```python
# Contar e calcular propor√ß√µes
gender_counts = df['Genero'].value_counts()
gender_prop = (gender_counts / gender_counts.sum() * 100)

# Arredondar manualmente: >=0.6 sobe para o inteiro mais pr√≥ximo
gender_prop = gender_prop.apply(lambda x: int(x) + 1 if (x % 1) >= 0.6 else int(x))

# Ajustar soma para 100% (caso o arredondamento passe de 100)
diff = 100 - gender_prop.sum()
if diff != 0:
    # Corrige distribuindo o ajuste ao maior valor (ou menor, se sobrar)
    idx = gender_prop.idxmax() if diff < 0 else gender_prop.idxmin()
    gender_prop[idx] += diff

# Criar tabela resumo
gender_table = pd.DataFrame({
    'G√™nero': gender_counts.index,
    'Quantidade': gender_counts.values,
    'Propor√ß√£o (%)': gender_prop
})

# Criar figura com 2 partes (tabela + gr√°fico)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Exibir tabela
axes[0].axis('tight')
axes[0].axis('off')
table = axes[0].table(
    cellText=gender_table.values,
    colLabels=gender_table.columns,
    cellLoc='center',
    loc='center'
)
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.1, 1.3)

# Ajuste autom√°tico de largura das colunas conforme o texto
col_widths = [max(gender_table[col].astype(str).map(len).max(), len(col)) * 0.015 for col in gender_table.columns]
for i, width in enumerate(col_widths):
    for (row, col), cell in table.get_celld().items():
        if col == i:
            cell.set_width(width)

# Cores personalizadas
colors = ['#4C72B0', '#DD8452', '#55A868']

# Criar gr√°fico de rosca (donut)
axes[1].pie(
    gender_counts,
    labels=gender_counts.index,
    autopct=lambda p: f'{int(round(p))}%',
    startangle=90,
    colors=colors[:len(gender_counts)],
    wedgeprops={'width': 0.6, 'edgecolor': 'white'}
)
axes[1].set_title('Distribui√ß√£o por G√™nero', fontsize=11)

# T√≠tulo do Quadro
fig.suptitle('Propor√ß√£o de estudantes por g√™nero', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 4.2 üßÆ Estat√≠sticas descritivas das idades

```python
# Fun√ß√£o para calcular estat√≠sticas descritivas
def stats_summary(data):
    return {
        'M√≠nimo': data.min(),
        'M√°ximo': data.max(),
        'M√©dia': round(data.mean(), 1),
        'Mediana': data.median(),
        'Vari√¢ncia': round(data.var(), 2),
        'Desvio Padr√£o': round(data.std(), 2)
    }
```

```python
# Criar dicion√°rios de estat√≠sticas
geral = stats_summary(df['Idade'])
masc = stats_summary(df[df['Genero'] == 'Homem']['Idade'])
fem = stats_summary(df[df['Genero'] == 'Mulher']['Idade'])
nb = stats_summary(df[df['Genero'] == 'N√£o bin√°rio']['Idade'])
```

```python
# Converter para DataFrames
stats_geral = pd.DataFrame([geral])
stats_masc = pd.DataFrame([masc])
stats_fem = pd.DataFrame([fem])
stats_nb = pd.DataFrame([nb])
```

```python
# Criar figura com 4 linhas
fig, axes = plt.subplots(4, 1, figsize=(8, 8))
fig.suptitle('Estat√≠sticas Descritivas das Idades\n', fontsize=14, fontweight='bold')

# Fun√ß√£o para exibir tabela com fundo alternado
def show_table(ax, df, title, color):
    ax.axis('tight')
    ax.axis('off')
    table = ax.table(
        cellText=df.values,
        colLabels=df.columns,
        cellLoc='center',
        loc='center'
    )

    # Estilos da tabela
    table.auto_set_font_size(False)
    table.set_fontsize(9)
    table.scale(1.1, 1.3)

    # Cor do t√≠tulo
    ax.set_title(title, fontsize=12, fontweight='bold', color=color)

    # Aplicar fundo alternado nas c√©lulas
    for (row, col), cell in table.get_celld().items():
        if row == 0:  # cabe√ßalho
            cell.set_text_props(weight='bold', color='white')
            cell.set_facecolor('#4F4F4F')
        elif row % 2 == 0:
            cell.set_facecolor('#F2F2F2')
        else:
            cell.set_facecolor('#FFFFFF')
        cell.set_edgecolor('#C0C0C0')

# --- Exibir as tabelas empilhadas ---
show_table(axes[0], stats_geral, "Geral", "#333333")
show_table(axes[1], stats_masc, "Masculino", "#4C72B0")
show_table(axes[2], stats_fem, "Feminino", "#C44E52")
show_table(axes[3], stats_nb, "N√£o Bin√°rio", "#8172B3")

plt.tight_layout()
plt.show()
```

```python
# Fun√ß√£o auxiliar para remover acentos e pontua√ß√£o
def clean_text(texto):
    # Remover acentos
    texto = ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn'
    )
    # Remover pontos e v√≠rgulas extras
    texto = re.sub(r'[.,;:!?]+$', '', texto.strip())
    return texto

# Limpeza e separa√ß√£o das respostas abertas
subjects_series = (
    df['Materia_Preferida']
    .dropna()
    .str.replace(' e ', ',', regex=False)
    .str.replace('/', ',', regex=False)
    .str.split(',')
)

# Cria lista limpa, sem acentos e sem pontua√ß√£o
subjects_list = [
    clean_text(s.strip().capitalize())
    for sublist in subjects_series
    for s in sublist
    if s.strip() != ''
]

# Contagem de ocorr√™ncias
subjects_counts = pd.Series(subjects_list).value_counts().reset_index()
subjects_counts.columns = ['Materia', 'Quantidade']

# Criar figura (tabela + gr√°fico lado a lado)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
fig.subplots_adjust(wspace=0.4)

# Tabela
axes[0].axis('tight')
axes[0].axis('off')
table = axes[0].table(
    cellText=subjects_counts.head(12).values,
    colLabels=subjects_counts.columns,
    cellLoc='center',
    loc='center'
)
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.1, 1.3)

# Gr√°fico de barras horizontal
axes[1].barh(subjects_counts['Materia'][:12], subjects_counts['Quantidade'][:12], color='#4C72B0')
axes[1].invert_yaxis()  # mat√©ria mais popular no topo
axes[1].set_xlabel('Quantidade de alunos')
axes[1].set_title('Materias preferidas', fontsize=12, fontweight='bold')

plt.suptitle('Materias preferidas (resposta aberta)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
