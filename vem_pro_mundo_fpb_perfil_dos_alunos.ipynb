## 4.1 üöª Propor√ß√£o de estudantes por g√™nero

```python
# Calcular contagem e propor√ß√£o
gender_counts = df['Com qual g√™nero voc√™ se identifica?'].value_counts()
gender_prop = (gender_counts / gender_counts.sum() * 100).round(0).astype(int)

# Ajustar para fechar em 100%
diff = 100 - gender_prop.sum()
gender_prop[gender_prop.idxmax()] += diff

# Criar DataFrame
gender_stats = pd.DataFrame({
    'G√™nero': gender_counts.index,
    'Quantidade': gender_counts.values,
    'Propor√ß√£o (%)': gender_prop.values
})

# Gerar Markdown
markdown_table = "| G√™nero | Quantidade | Propor√ß√£o (%) |\n|---------|-------------|----------------|\n"
for i, row in gender_stats.iterrows():
    markdown_table += f"| {row['G√™nero']} | {row['Quantidade']} | {row['Propor√ß√£o (%)']}% |\n"

# Exibir tabela
display(Markdown(markdown_table))
```

```python
# Gr√°fico de pizza
plt.figure(figsize=(6,6))
plt.pie(gender_prop, labels=gender_counts.index, autopct='%1.0f%%', startangle=90,
        colors=['#66b3ff','#ff9999','#99ff99'])
plt.title('Distribui√ß√£o de G√™nero')
plt.show()
```

## 4.2 üßÆ Estat√≠sticas descritivas das idades

```python
# Calcular as estat√≠sticas
age_stats = pd.DataFrame({
    'M√≠nimo': [df['Idade'].min()],
    'M√°ximo': [df['Idade'].max()],
    'M√©dia': [df['Idade'].mean()],
    'Mediana': [df['Idade'].median()],
    'Vari√¢ncia': [df['Idade'].var()],
    'Desvio padr√£o': [df['Idade'].std()]
})

# Arredondar valores
age_stats = age_stats.round(2)
```

```python
# Transpor para formato vertical
age_stats_vertical = age_stats.T.reset_index()
age_stats_vertical.columns = ['Estat√≠stica', 'Valor']

# Gerar string Markdown automaticamente
markdown_table = "| Estat√≠stica | Valor |\n|--------------|-------|\n"
for i, row in age_stats_vertical.iterrows():
    markdown_table += f"| {row['Estat√≠stica']} | {row['Valor']} |\n"

# Exibir tabela
display(Markdown(markdown_table))
```

```python
# Calcular estat√≠sticas descritivas da idade por g√™nero
age_stats_gender = df.groupby('Com qual g√™nero voc√™ se identifica?')['Idade'].agg(
    M√≠nimo='min',
    M√°ximo='max',
    M√©dia='mean',
    Mediana='median',
    Vari√¢ncia='var',
    Desvio_Padr√£o='std'
).round(2).reset_index()
```
